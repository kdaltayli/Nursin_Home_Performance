#### Plot a bar chart showing the number of nursing homes in states across US

```python
# Function to draw bar chart
# Statewise Count of Nursing Homes in US
def drawBar_CountAllStates(provider_df):
    prov_groupby_state = provider_df.groupby("State")['Prov Name'].count().reset_index(name="count by state")
    prov_groupby_state = prov_groupby_state.sort_values('count by state', ascending=False)
    x_values = prov_groupby_state['State']
    y_values = prov_groupby_state['count by state']
    
    plt.figure(figsize=(32,12))
    plt.grid(True)
    plt.bar(x_values, y_values) 
    plt.title("Nursing Homes Count in United States", fontsize=30)
    plt.xlabel("States", fontsize=24)
    plt.ylabel("Number of Nursing Homes", fontsize=24)    
    ax = plt.gca()
    ax.tick_params(axis = 'x', which = 'major', labelsize = 20,labelrotation =75)    
    ax.tick_params(axis = 'y', which = 'major', labelsize = 20)    
#     plt.style.use('seaborn')    
    plt.savefig("../Output/nhallstates.png")
    plt.show()
```

#### Plot a stacked bar chart with top ten best rated nursing homes in US

```python
# Function to draw stacked bar plot.
# Top ten states having best rated nursing homes
def drawBar_TopTenAllStates(provider_info):
    # Retrieve the data fro the top ten providers
    nh_rated_topten = getToptenNames(provider_info)        
    nh_rated_topten.plot(x='State',  y=["Total", "Top Rated Total"], kind='bar', stacked=True)
    plt.xlabel("States")
    plt.ylabel("Number of Nursing Homes")
    plt.legend(["Total Nursing Homes", "Best Rated Nursing Homes"])
    plt.title("Top Ten States in US having Best Rated Nursing Homes")
    plt.grid(True)
#     plt.style.use('bmh')
    plt.savefig("../Output/nhtop10states.png")
    plt.show()
```

#### Plot a line chart showing the various ratings that contribute to the overall rating of the nursing homes

```python
# Overall Rating of the nursing homes vs Other Ratings
def drawOverallvsOtherRatings():
    providerinfo_data = getProviderRatings()
    groupby_rating = providerinfo_data.groupby('Overall_Rating').mean().reset_index()

    y_axis = groupby_rating['Overall_Rating']
    x_axis_1 = groupby_rating['SURVEY_RATING']
    x_axis_2 = groupby_rating['Quality_Rating']
    x_axis_3 = groupby_rating['Staffing_Rating']
    x_axis_4 = groupby_rating['RN_staffing_rating']
    x_axis_5 = groupby_rating['LS_Quality_Rating']
    x_axis_6 = groupby_rating['SS_Quality_Rating']
    ratings = ["Health Inspection Rating", "Quality Rating", "Staff Rating", "Registered Nurse Rating", 'Long Stay Quality Rating', 'Short Stay Quality Rating']

    plt.plot(x_axis_1, y_axis, linewidth=1, marker="o", color="r")
    plt.plot(x_axis_2, y_axis, linewidth=1, marker="s", color="g")
    plt.plot(x_axis_3, y_axis, linewidth=1, marker="*", color="b")
    plt.plot(x_axis_4, y_axis, linewidth=1, marker="^", color="y")
    plt.plot(x_axis_5, y_axis, linewidth=1, marker="^", color="m")
    plt.plot(x_axis_6, y_axis, linewidth=1, marker="^", color="c")
    plt.title("Other Ratings vs Overall Ratings")
    plt.ylabel("Overall Ratings")
    plt.xlabel("Other Rating")
    plt.legend(ratings)
    plt.grid(True)
#     plt.style.use('bmh')
    plt.savefig("../Output/ratings.png")
    plt.show()
```

#### Plotting the data about the Centenarians in PA nursing homes

```python
def residentsOver100(pennstate_census_data):
#     pennstate_census_data = getPAResidentCensus()
    total_men = pennstate_census_data['Men above 100'].sum()
    total_women = pennstate_census_data['Women above 100'].sum()

    genderwise_df = pd.DataFrame({
        'Gender' : ['Male', 'Female'],
        'Count' : [total_men, total_women]
    })
    genderwise_df

    labels = [ f'Men : {total_men}' , f'Women : {total_women}']
    colors = ['aqua', 'hotpink']
    plt.title("PA Nursing Home Residents Aged above 100")  #, bbox={'facecolor':'0.8', 'pad':5})
    plt.pie(genderwise_df['Count'], labels=labels, colors=colors, autopct="%1.1f%%", startangle=120,  explode=(0,0.06), shadow=True)
    plt.axis("equal")
#     plt.style.use('classic')
    plt.grid(True)
    plt.savefig("../Output/Over100.png")
    plt.show()
```

#### Plot agewise distribution of residents in PA nursing homes

```python
def drawBar_AgewiseResidentsPA(pennstate_census_data):
    temp_agewise_df = pennstate_census_data[['Resident below 18', 'Resident 18 to 44', 'Resident 45 to 59', 
                                                  'Resident 60 to 64', 'Resident 65 to 69', 'Resident 70 to 74',
                                                  'Resident 75 to 79', 'Resident 80 to 84' , 'Resident 85 to 89', 
                                                  'Resident 90 to 94', 'Resident 95 to 99', 'Resident above 100']]
    
    temp_agewise_df = temp_agewise_df.rename(columns={'Resident below 18' : 'Below 18', 'Resident 18 to 44' : '18 to 44', 'Resident 45 to 59' : '45 to 59',
                        'Resident 60 to 64' : '60 to 64', 'Resident 65 to 69' : '65 to 69', 'Resident 70 to 74' : '70 to 74',
                        'Resident 75 to 79' : '75 to 79', 'Resident 80 to 84' : '80 to 84', 'Resident 85 to 89' : '85 to 89',
                        'Resident 90 to 94' : '90 to 94', 'Resident 95 to 99' : '95 to 99', 'Resident above 100' : 'Above 100'})
    agewise_resident_df = temp_agewise_df.sum()

    plt.rcdefaults()
    title = "Agewise Number of Residents in PA Nursing Homes"
    plt.xlabel('Age Groups', fontsize=16)
    plt.ylabel('Number of residents', fontsize=16)
    plt.title(title, fontsize=16)
    agewise_resident_df.plot(kind='bar', figsize=(20,10), color='g', alpha=0.8)
    ax = plt.gca()
    ax.tick_params(axis = 'x', which = 'major', labelsize = 13, labelrotation =75)
    ax.tick_params(axis = 'y', which = 'major', labelsize = 13)
#     plt.style.use('bmh')
    plt.grid(True)
    plt.savefig("../Output/agewise")
    plt.show()
```

#### Plot the ten PA counties with highest covid death counts

```python
def drawBar_covid19TopTenUSStates(covid_19_us):
    # Get covid death rates by state
    clear_covid_19 = covid_19_us.groupby("STATE").sum()
#     clear_covid_19["Percentage"] = round((clear_covid_19["Residents Total COVID-19 Deaths"]/clear_covid_19["Residents Total All Deaths"])*100,3)
    
    # Get the highest 10 states with covid-19 death rate
    Covid_19_death_sorted = clear_covid_19.sort_values(by='Residents Total COVID-19 Deaths', ascending=False)
    
    # Worst 10 states across US based on the COVID-19 Deaths 
    top_ten = Covid_19_death_sorted.head(10).reset_index()    
    
    # Graph Covid_19_Total Number of Covid-19 Deaths in Worst Top Ten States across US
    top_ten.plot(kind="bar",x="STATE", y=["Residents Total All Deaths","Residents Total COVID-19 Deaths"],)
    plt.title("Ten States with Most COVID-19 Deaths")
    plt.ylabel("Death Counts in Nursing Homes")
    ax = plt.gca()
    ax.invert_yaxis()
    ax.tick_params(axis = 'x', which = 'major', labelsize = 12, labelrotation =75)
    plt.grid(True)
    plt.style.use('seaborn')
    
#     plt.tight_layout()
    plt.savefig("../Output/covidtopten.png")
    plt.show()
```

#### Plot the ten PA counties with least covid death counts

```python
def drawBar_covid19LastTenUSStates(covid_data_us):
    # Get covid death rates by state
    clear_covid_19 = covid_data_us.groupby("STATE").sum()
    
    # Get the last 10 states with covid-19 death rate
    Covid_19_death_sorted = clear_covid_19.sort_values(by='Residents Total COVID-19 Deaths', ascending=False)
    
    # Best 10 states across US based on the COVID-19 Deaths 
    last_ten = Covid_19_death_sorted.tail(10).reset_index()
    
    # def covid_19_Total Number of Covid-19 Deaths in Best Top Ten States across US
    last_ten.plot(kind="bar",x="STATE", y=["Residents Total All Deaths","Residents Total COVID-19 Deaths"])
    plt.grid(False)
    plt.title("Ten States with Least COVID-19 Deaths")
    ax = plt.gca()
    ax.tick_params(axis = 'x', which = 'major', labelsize = 12, labelrotation =75)
    plt.ylabel("Death Count in Nursing Homes")
    plt.style.use('bmh')
    plt.grid(True)
#     plt.tight_layout()
    plt.savefig("../Output/covidlastten.png")
    plt.show()
```

#### Plot the toal beds vs occupied beds in nursing homes

```python
def drawBar_NumberofBeds(plotting_df, subtitle, saveasfile):
    y2 = plotting_df['Total Number of Occupied Beds']   # cyan
    y1 = plotting_df['Number of All Beds']  # yellow
    x = plotting_df['Provider State']

    # Graph Covid_19_Total Number of Covid-19 Deaths in Worst Top Ten States across US
    plotting_df.plot(kind="bar",x="Provider State", y=['Number of All Beds', 'Total Number of Occupied Beds'], figsize=(20,10), color=['b','g'])
    plt.title(f"Number of Beds vs Occupied Beds Across US\n {subtitle}", fontsize=24)
    plt.ylabel("Number of Beds (Total vs Occupied)", fontsize=20)
    plt.xlabel("State", fontsize=20)
    plt.legend(fontsize=16)
    ax = plt.gca()
    ax.tick_params(axis = 'x', which = 'major', labelsize = 16, labelrotation =75)
    ax.tick_params(axis = 'y', which = 'major', labelsize = 16)
    plt.savefig(f"../Output/{saveasfile}.png")
    plt.grid(True)
    plt.style.use('bmh')
```

```python
#Read the input data-1/1/2020- 7/16/2020
def getCovid19AcrossUS():
    # Read the data from the csv file
    read_covid_19 = pd.read_csv("../Resources/COVID-19_Nursing_Home_Dataset.csv")
    
    # Retrieve the relevant columns from the dataset
    covid_19 = read_covid_19[["Provider Name","Provider State","Residents Total All Deaths","Residents Total COVID-19 Deaths","County","Geolocation"]]

    # Rename the ProviderNAme and State
    renamed_covid_19 = covid_19.rename(columns = {"Provider Name":"Prov Name","Provider State":"STATE"})

    # Drop the records with Residents total Covid-19 Deaths zero
    renamed_covid_19.drop(renamed_covid_19[(renamed_covid_19["Residents Total All Deaths"] == 0) | (renamed_covid_19["Residents Total All Deaths"].isnull())].index, inplace = True)

    return (renamed_covid_19)
```

## Analysis

*  Texas, California, Ohio, Illinois and Pennsylvania are the top 5 States with best rated Nursing Homes.

*  Over-all rating is an average of long-stay /short stay rating, survey rating, skills

*  Nursing homes with more nurses, orderlies tend to be slightly better rated.

*  Irrespective of the Nursing Homes rating, ratio of women living in Nursing home is 40% more when compared to Men.

*  Ratio of Women vs Men living in Nursing Home is 7:3

*  In PA, around 18% of the total residents in Nursing Homes are of age group 85 to 89, followed by 16% of age group 90 and 95

*  An interesting fact is that there are around 1000 people in PA nursing homes who are aged more than 100 years.

*  In the current situation we see that the nursing homes with more people show more deaths due to COVID than the ones with fewer residents.

*  Shows the need for more nursing homes in states like NH, VT, ME with greater aging population.

### Prediction

As we age, daily activities tend to become more challenging, especially if chronic health conditions like diabetes, arthritis are also present.

*  Based on Medicaid & Medicare services report, In US 33.8% of residents in Nursing Homes were 85 to 94 years old, 26.4% were 75 to 84 years old and 16.5% were 65 to 74 years old.

From our analysis, we strongly believe that nursing homes in some states with greater populations are crowded.

*  If the nursing home needs increase by 5% then the already crowded Nursing Homes are not equipped to provide the care our seniors require. Not to mention if another pandemic like the one we are facing now strikes again.

*  In Nursing Homes across US, an average of 80% Beds are occupied leaving shortage of Beds if needed (also if there is a wave 2)
